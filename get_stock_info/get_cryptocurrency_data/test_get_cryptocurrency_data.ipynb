```python
import cryptocompare
import datetime
import mplfinance as mpf
import pandas as pd
```

```python
coin_list = cryptocompare.get_coin_list(format=False)
for key, value in coin_list.items():
    print("coin: {}".format(key))
```

```python
data_raw = cryptocompare.get_historical_price_day('BTC', 'JPY', limit=150, exchange='CCCAGG', toTs=datetime.datetime(2020,12,19))
data_df = pd.DataFrame(data_raw)
data_df["timestamp"] = pd.to_datetime(data_df['time'], unit='s')
data_df = data_df.set_index("timestamp")
print(data_df.index)
mpf.plot(data_df,type='candle', mav=(7, 15, 30))
```
